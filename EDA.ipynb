## 4. Data visualization



```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

## 5. Hypothesis investigation

Hypothesis: The client's hypothesis is that the price sensitivity has influence on churn.

We define the price sensitivity as, $\textrm{price sensitivity}=\frac{\textrm{consumption change}}{\textrm{price change}}$.
Based on the equation, we process the data in `price_df`.




```python
ite_df = pd.DataFrame({})
price_col = ['price_off_peak_var','price_peak_var','price_mid_peak_var','price_off_peak_fix','price_peak_fix','price_mid_peak_fix']
for company in client_df['id'].unique():
  temp_df = price_df[price_df['id']==company] 
  if len(temp_df)==1:
    price_var = temp_df[price_col]
  else:
    time_max = pd.to_datetime(temp_df['price_date']).idxmax()
    time_min = pd.to_datetime(temp_df['price_date']).idxmin()
    price_var = temp_df.loc[time_max,price_col] - temp_df.loc[time_min,price_col] # the price change over the whole year
    price_var_df = pd.DataFrame({'id':[company]}) 
    price_var_df[price_col] = price_var.values
    ite_df = ite_df.append(price_var_df,ignore_index=True)
```

```python
# Merge consumption with ite_df
cons_price = consumption.merge(ite_df,on='id')
```

* The distributions of price changes are close to the bell-shape normal distribution.

```python
nrow = len(cons_price_df['cons_last_month'].to_numpy())
```

```python
sen_col = ['sen_off_peak_var','sen_peak_var','sen_mid_peak_var','sen_off_peak_fix','sen_peak_fix','sen_mid_peak_fix']
```

```python
sensitivity = np.repeat(np.reshape(cons_price_df['cons_last_month'].to_numpy()-cons_price_df['imp_cons'].to_numpy(),(nrow,1)),6,axis=1)/(cons_price_df[price_col].to_numpy())
cons_price_df[['sen_off_peak_var','sen_peak_var','sen_mid_peak_var','sen_off_peak_fix','sen_peak_fix','sen_mid_peak_fix']] = sensitivity
```

```python
corr = cons_price_df[price_col+sen_col+['churn']].corr()

fig, ax = plt.subplots(figsize=(18, 18))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot = True, cmap='YlGnBu')
ax.set_title('Correlation heatmap')
```

* From the correlation heatmap, we observe that the linear dependence between churn and price sensitivity is small, which requires more sophisticated feature extraction processing.

## Summary

Through EDA, we conclude the following findings:
1. The data is unbalance in `churn`, with only 9.7% of companies churn within the next three months.
2. The distributions of consumptions and forecasts are skewed.
3. The numbers of outliers in consumptions and forecasts are not negligible.
4. Clients with 2 years' subscription are more likely to churn while clients with more-than-4-year subscription have churn ratios below 10%.
5. The distributions of price sensitivity are analogous to normal distribution.
6. The linear independence between `churn` and other variables, including price sensitivity, is small.
7. There exists relationship between `churn` and other categorical data.

Based on the findings, several suggestions are provided:
1. The influence of unbalanced data and skewed distribution needs to be addressed. For some data, log-scale operation can make it close to normal distribution, which may ease these influences.
2. Feature engineering is desired to be employed to extract features to construct efficient predictive learning models.
3. The definition of price sensitivity could be improved.
4. More data sets could be included, e.g., the comments from served companies especially the churning ones, price from other peer energy suppliers, the financial status of churning companies, etc.
